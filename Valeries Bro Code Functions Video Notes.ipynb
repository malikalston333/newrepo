```python
# # Defining and Calling a Function
# To define a function in Python:
# Use the def keyword followed by the function name and parentheses.
# Add a colon : at the end.
# Indent the block of code that belongs to the function.




def happy_birthday():
    print("Happy birthday to you!")


# To run this function, simply type:


happy_birthday()

# instead of repeating lines of code or using loops, we can create functions!
# You can call this function as many times as needed:


happy_birthday()
happy_birthday()
happy_birthday()
```

```python
# Passing Arguments to Functions
# Functions become more powerful when they accept arguments
# Arguments are values you pass into them for customization.

# this function has 1 parameter that accepts 1 argument
def happy_birthday(name):
    print(f"Happy birthday to {name}!")


# Now, we can call it with different names:

happy_birthday("Bro")
happy_birthday("Valerie")
happy_birthday("Hope")
```

```python
# a function with multiple parameters

def happy_birthday(name, age):
    print(f"Happy birthday to {name}!")
    print(f"You are {age} years old.")

# Invoking the function with multiple arguments by passing 2 arguments inside the parenthses
happy_birthday("Bro", 20)

# pass data values or variables to the function.
happy_birthday("Steve", 30)
happy_birthday("Joe", 40)


#dont mess up the order.... order matters in this example.
happy_birthday(20, "Bro")


#we dont have to use matching strings like "name" when we want to create a function with a parameter
#here we just use x and y. but this isnt very helpful for the next person to read your code
def happy_birthday(x, y):
    print(f"Happy birthday to {x}, you are {y} years old.")

```

```python
# Example 2: Invoice Function with 3 parameters

def display_invoice(username, amount, due_date):
    print(f"Hello, {username}")
    print(f"Your bill of ${amount:.2f} is due on {due_date}")  #the .2f means 2 decimal places so we dont get crazy long cents

# Invoking the invoice function
display_invoice("Valerie Boss", 1542.52654780, "January 1")
display_invoice("Joe Schmoe", 100.01, "February 1")

```

```python
# Example 3: Math Functions Using Return instad of just printing to the screen
# a return sends a value back from the function to where it was called.

def add(x, y):
    z = x + y
    return z

def subtract(x, y):
    z = x - y
    return z

def multiply(x, y):
    z = x * y
    return z

def divide(x, y):
    z = x / y
    return z

# Using the math functions
print(f"Add: {add(1, 2)}")        # Output: 3
print(f"Subtract: {subtract(1, 2)}")  # Output: -1
print(f"Multiply: {multiply(1, 2)}")  # Output: 2
print(f"Divide: {divide(1, 2)}")      # Output: 0.5

# Example 4: Create Full Name Function
#capitalize first name adn capitalize last name then combine

def create_name(first, last):
    first = first.capitalize() #capitalize is a string method
    last = last.capitalize()
    return f"{first} {last}"

# Using the create_name function
#pass fist name and passname to the function.  assign the returne value to a variable full_name then print it.
full_name = create_name("john", "doe")
print(full_name)  # Output: John Doe


#i wonder if i can pass VARIABLES into the function...
column1 = "Valerie"  #this is a global variable because it is outside the function
#mind blown"  inside the function, youâ€™re not directly using the global variables. You are passing their values as arguments to the function!

full_name = create_name(column1, "squarepants")
print(full_name)  # Output: Spongebob Squarepants

# is this what pass by value vs pass by reference means??
#***one eternity later***  and 12 articles ....nope.....
# Pass by Value means A copy of the data is passed into the function.  The function works with the copy and does not affect the original variable outside the function.

# Pass by Reference means A reference (or pointer) to the original data is passed into the function.  Changes made to the parameter can affect the original variable outside the function.

#im gonna need to look into this more....  :) how does python handle variables?

# ### Video examples summary of Concepts Covered:

# 1. **Function Definition**: Using `def` to define reusable blocks of code.
# 2. **Function Invocation**: Calling a function by its name followed by `()`.
# 3. **Arguments & Parameters**:
#    - *Arguments* are the values you pass in when calling the function.
#    - *Parameters* are placeholders for those values inside the function definition.
# 4. **Using Return**: Functions can return values to be used later.
# 5. **String Formatting**:
#    - `f"..."` for f-strings.
#    - `:.2f` to format floating-point numbers to two decimal places.
# 6. **Best Practices**: Reusing code, avoiding repetition, and using descriptive names.
```
